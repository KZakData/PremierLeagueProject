```python
def scrap1_premier_league_stats():
    
    from bs4 import BeautifulSoup
    import requests

    url='https://fbref.com/en/comps/9/Premier-League-Stats'

    page = requests.get(url)

    soup = BeautifulSoup(page.text,'html')
    
    soup.find_all('table')[1]
    
    table = soup.find("table", class_ ="stats_table sortable min_width force_mobilize")
    
    table_titles=table.find_all('th')
    
    all_table_titles = [title.text for title in table_titles][1:20]
    
    import pandas as pd
    
    df = pd.DataFrame(columns = all_table_titles)
    
    column_data = table.find_all('tr')
    
    for row in column_data[1:]:
        row_data = row.find_all('td')[:19]
        individual_row_data = [data.text.strip() for data in row_data]
        lenght = len(df)
        df.loc[lenght]=individual_row_data
        
    df['Squad'].replace('Newcastle Utd','Newcastle',inplace = True)
    df['Squad'].replace('Sheffield Utd','Sheffield',inplace = True)
    df['Squad'].replace(df.iloc[12][0],'Nottingham Forest',inplace = True)
    df['Squad'].replace('Luton Town','Luton',inplace = True)
    df['xG'] = df['xG'].str.replace(r'\.', ',', regex=True)
    
    df.to_csv(r'C:\Users\zakv\Desktop\Data_analyst\PremierLeaguexG.csv',  index=False)
    
    print(df)
```

```python
scrap1_premier_league_stats()
```

```python
def scrap2_premier_league_stats():
    
    from bs4 import BeautifulSoup
    import requests
    
    url='https://www.topscorersfootball.com/premier-league'

    page = requests.get(url)

    soup = BeautifulSoup(page.text,'html')
    
    soup.find_all('table')[1]
    
    table = soup.find("table", class_ ="table table-striped")
    
    table_titles=table.find_all('th')
    
    all_table_titles = [title.text for title in table_titles]
    
    import pandas as pd
    
    df = pd.DataFrame(columns = all_table_titles)
    
    column_data = table.find_all('tr')
    
    for row in column_data[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        lenght = len(df)
        df.loc[lenght]=individual_row_data

    df['Team'].replace('Sheffield United','Sheffield',inplace = True)
    df['Team'].replace('Manchester United','Manchester Utd',inplace = True)
    
    df.to_csv(r'C:\Users\zakv\Desktop\Data_analyst\PremierLeagueTopScores.csv',  index=False)
    
    print(df)
```
